```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn import svm

#Reading data from csv file
data = pd.read_csv('titanic.csv')

#Predicting survivor column
columns_target = ['Survived']

#Prediction based on following 4 columns data
columns_train = ['Age','Pclass','Sex','Fare']

X = data[columns_train]
Y = data[columns_target]

#Checking which fields have NaN values 

X['Sex'].isnull().sum()
X['Pclass'].isnull().sum()
X['Fare'].isnull().sum()
X['Age'].isnull().sum()

#Replacing all NaN values of Age with median of the field

X['Age'] = X['Age'].fillna(X['Age'].median())

#Field Sex is a categorical field. Converting to 0 or 1 for prediction
d = {'male':0,'female':1}
X['Sex'] = X['Sex'].apply(lambda x:d[x])



X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.33, random_state = 42)
clf = svm.LinearSVC()

clf.fit(X_train,Y_train)

print(clf)


#Predicting values for first 10 imputs
print(clf.predict(X_test[0:10]))

# Checking accuracy of the model
print(clf.score(X_test,Y_test))
```
